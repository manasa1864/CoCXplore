### üîº **Upgrading Pokemon Stats**
Sometimes, Pok√©mon need a power-up! The `upgrade_stats()` function **squares** all Pok√©mon stats, making them stronger.  
- `np.square(stats)`: Squares each value (e.g., `10 ‚Üí 100`).  
- `np.multiply(stats, stats)`: Also squares values but demonstrates an alternative approach.  

üí° Run the function and see how the Pok√©mon becomes stronger!

```python
import numpy as np
duplicate_df = df.copy()

def upgrade_stats(pokemon_name):
    """Upgrades a specific Pok√©mon's stats using NumPy's square and multiply operations."""
    if pokemon_name not in df["name"].values:
        print(f"‚ö†Ô∏è Pok√©mon '{pokemon_name}' not found in the database.")
        return

    print(f"\n Upgrading Stats for {pokemon_name} üîº üîº\n")

    # Locate the Pok√©mon's row index
    index = df[df["name"] == pokemon_name].index[0]

    # Select the Pok√©mon's stats
    stats = df.loc[index, ["hp", "attack", "defense", "speed"]]
    special_stats = df.loc[index, ["sp_attack", "sp_defense"]]

    # Upgrade physical stats using np.square (element-wise squaring)
    df.loc[index, ["hp", "attack", "defense", "speed"]] = np.square(stats)

    # Upgrade speacial stats using np.multiply
    df.loc[index, ["sp_attack", "sp_defense"]] = np.multiply(special_stats, 2)

    print(f"‚úÖ Stats upgraded for {pokemon_name}!")
    print(df.loc[index, ["name", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]])

pokemon_name = input("Enter Pok√©mon name to upgrade (Ex: Pikachu): ").strip()
upgrade_stats(pokemon_name)
```

### üîÑ **Reset Pokemon Stats**
Oops! Did you overpower your Pok√©mon too much? No worries `reset_stats()` restores all Pok√©mon to their original stats.


```python
def reset_stats():
    """Resets the Pok√©mon stats to their original values."""
    global df
    df = duplicate_df.copy()
    print("\nüîÑ All Pok√©mon stats have been reset to their original values!")

reset = input("Do you want to reset stats? (yes/no): ").strip().lower()
if reset == "yes":
    reset_stats()
```

### üé≤ **The Ultimate Random Battle!**
Want a surprise? `random_battle()` randomly selects two Pok√©mon and makes them fight!  
Try running it multiple times‚Äîevery battle is different!

```python
def random_battle():
    """Randomly selects two Pok√©mon and simulates a battle."""
    pokemon1, pokemon2 = np.random.choice(df["name"].tolist(), 2, replace=False)
    print(f"\nüéÆ Battle: {pokemon1} vs {pokemon2}")
    battle(pokemon1, pokemon2)

random_battle()
```
