## **2. Initializing Arrays**

```python
# Creating an Array of Zeros
zeros_array = np.zeros((3, 3))
print("Zeros Array:\n", zeros_array)
```

```python
# Creating an Array of Ones
ones_array = np.ones((2, 4))
print("Ones Array:\n", ones_array)
```

```python
# Generating a Random Array
random_array = np.random.random((3, 2))
print("Random Array:\n", random_array)
```

```python
# Find shape of array
import numpy as np
A=[[1,2,3],[4,5,6]]
B=[[1,2],[5,3],[4,5]]
def is_compatible(A, B):
    r1=0
    c1=0
    r2=0
    c2=0
    cnt=0
    for i in A:
        if cnt==0:
          for j in i:
            c1+=1

        r1+=1
        cnt+=1
    for i in B:
        if cnt==r1:
          for j in i:
            c2+=1

        r2+=1
        cnt+=1
    if c1==r2:
        print("The dot product is: ",np.dot(A,B))
    else:
        print("Dimension Error!!")
is_compatible(A,B)
```

## **3. Reshaping of Arrays**

```python
# Reshaping a 1D array to 2D
print("Before Reshaping:\n", array1)
reshaped_array = array1.reshape(3, 2) # a1.reshape(rows, columns)
print("After Reshaping:\n", reshaped_array)
```


## **4. Summation and Average**

```python
# Summing elements
print("Array:\n", array2)
print("\nSum of elements in an array: ", np.sum(array2))
```

```python
# Computing the average
print("Array:\n", array2)
print("\nAverage of elements in an array", np.average(array2))
```


## **5. Dot Product in NumPy**

```python
# Creating two matrices
import numpy as np

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Display matrices
print("Matrix 1:\n", matrix1)
print("\nMatrix 2:\n", matrix2)

# Dot product
dot_product = np.dot(matrix1, matrix2)
print("\nDot Product:\n", dot_product)
```

## **6. Transpose of an Array**

```python
# Transposing matrix1
print("Matrix:\n", matrix1)
transpose_matrix = np.transpose(matrix1)
print("\nTranspose of matrix:\n", transpose_matrix)
```
